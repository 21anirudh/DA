```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Points
x = np.array([2, 3, 5, 7, 9])
y = np.array([1, 2, 4, 5, 6])

# Number of data points
n =len(x)
```

# SLR using Least Squares Method

```python
# Calculate sums required for least squares method
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_sy = np.sum(x*y)
sum_x_squared = np.sum(x**2)

# Calculate slope (m) and y-intercept (c)
m = (n*sum_sy - sum_x*sum_y) / (n*sum_x_squared - sum_x**2)
c = (sum_y - m*sum_x) / n

print(f"Slope (m): {m}")
print(f"y-Intercept (c): {c}")

# Plotting the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plotting the regression line
plt.plot(x, m*x + c, color='red', label=f'Regression Line: y = {m:.2f}x + {c:.2f}')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression using Least Squares Method')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

# SLR using Gradient Descent Method

```python
# Data Points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 8])

# Hyperparameters required for gradient descent method
learning_rate = 0.05
epochs = 200 # Number of iterations

# Initialize m (slope) and c (intercept) to zero
m = 0
c = 0

# Store history of m and c for plotting
m_history = [m]
c_history = [c]

# Gradient Descent Function
for epoch in range(epochs):
    # Calculate predicted values
    y_pred = m * x + c

    # Calculate the gradients
    dm = (-2/n) * np.sum(x * (y - y_pred))
    dc = (-2/n) * np.sum(y - y_pred)

    # Update parameters
    m -= learning_rate * dm
    c -= learning_rate * dc

    # Store updated parameters
    m_history.append(m)
    c_history.append(c)

    # Print values every 20 epochs
    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch + 1}: m = {m:.4f}, c = {c:.4f}")

# Plotting the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plotting the regression line
plt.plot(x, m * x + c, color='red', label=f'Regression Line: y = {m:.2f}x + {c:.2f}')

# Labeling the axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression using Gradient Descent Method')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

# Error Metrics

```python
def calculate_metrics(y_actual, y_pred, num_predictors):
    n = len(y_actual)

    # Mean of actual values
    y_mean = np.mean(y_actual)

    # Errors
    residuals = y_actual - y_pred

    # Metrics
    MAE = np.mean(np.abs(residuals))
    SSE = np.sum(residuals**2)
    MSE = np.mean(residuals**2)
    RMSE = np.sqrt(MSE)
    MAPE = np.mean(np.abs(residuals / y_actual)) * 100

    # R-squared
    SS_total = np.sum((y_actual - y_mean)**2)
    R_square = 1 - (SSE / SS_total)

    # Adjusted R-squared
    Adjusted_R_square = 1 - ((SSE / (n - num_predictors - 1)) / (SS_total / (n - 1)))

    # Print all metrics
    print(f"Mean Absolute Error (MAE): {MAE}")
    print(f"Sum of Squared Errors (SSE): {SSE}")
    print(f"Mean Squared Error (MSE): {MSE}")
    print(f"Root Mean Squared Error (RMSE): {RMSE}")
    print(f"Mean Absolute Percentage Error (MAPE): {MAPE}%")
    print(f"R-squared: {R_square}")
    print(f"Adjusted R-squared: {Adjusted_R_square}")

# Example usage:
y_actual = np.array([3, 4, 5, 6])
y_pred = np.array([2.5, 4.2, 4.8, 6.3])

# Assuming 1 predictor (simple linear regression)
num_predictors = 1

calculate_metrics(y_actual, y_pred, num_predictors)
```
